Distribution Plot

We use distribution plots to visualise the distribution of quantitative data

Let's discuss some plots that allow us to visualise the distribution of a dataset. These plots are:


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')

#Smoker, Sex are the categorical data or qualitative data
```

```python
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind='scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind='reg')
```

```python
sns.pairplot(df, hue = 'size',palette = 'rainbow')
```

```python
sns.rugplot(df['total_bill'])
```

## Categorical Data Plots


Now let's discuss using seaborn to plot categorical data. There are a few main plot types for this:

boxplot
violingplot
stripplot
swarmplot
barplot
countplot

```python
sns.countplot(x = df['sex'])
```

```python
sns.countplot(x = df['sex'], hue=df['smoker'])
```

```python
sns.barplot(x = df['sex'], y = df['total_bill'])
```

You can change the estimator object to your own function, that converts a vector to a vector

```python
import numpy as np 
sns.barplot(x = df['sex'], y=df['total_bill'], estimator=np.sum)
```

```python
import numpy as np 
sns.barplot(x = df['sex'], y=df['tip'], estimator=np.sum)
```

```python
sns.boxplot(x='tip',y = 'day', data = df)
```

## Stripplot and Swarmplot

THe stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its owm, but it  is also a good complement to a box or violin plot in cases where you want all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x='day',y = 'total_bill', data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill', data = df)
```

```python
sns.violinplot( x = 'tip',y = 'day', data = df)
sns.swarmplot( x = 'tip',y = 'day', data = df)
```

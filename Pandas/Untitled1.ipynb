```python
import numpy as np
import pandas as pd
```

Finding Missing data

```python
data = {
    'A':[1,2,np.nan,4,5],
    'B':[1,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5],
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

df

```python
df.dropna
```

```python
df.dropna(thresh=5)
```

Filling the missing data

```python
df.fillna(0)
```

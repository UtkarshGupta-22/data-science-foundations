```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean = 170
population_std = 3
samp_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (samp_mean - population_mean)/(population_std/np.sqrt(n))
```

```python
print(z_score)
```

```python
alpha = 0.05
```

```python
p_value = 2 * (1-norm.cdf(abs(z_score)))
```

```python
p_value
```

```python
if p_value < alpha :
    print("Reject null hypothesis")
else:
    print("I will accept null hypothesis")
```

```python
sample = [172, 174, 168,169,171,173,175,170,169,172]
sample_mean = np.mean(sample)
sample_std = np.std(sample,ddof=1)
n = len(sample)
mean_population = 170
```

```python
t_stats = (sample_mean - mean_population) / (sample_std / np.sqrt(n))
```

```python
t_stats
```

```python
p_value = 2 * (1 - stats.t.cdf(abs(t_stats), df = n-1))
alpha = 0.5
```

```python
if p_value < alpha:
    print("Reject NULL hypothesis")
else:
    print("Accept NULL Hypothesis")
```
